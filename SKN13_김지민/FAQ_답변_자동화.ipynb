```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
import time
import random
import pandas as pd

lst = []
keyword = input("검색어를 입력해주세요. :")
# 크롬 설정
options = Options()
options.add_argument("--disable-gpu")
options.add_argument("start-maximized")
options.add_argument("disable-infobars")
options.add_argument("--disable-blink-features=AutomationControlled")  # 자동화 감지 방지
options.add_experimental_option("excludeSwitches", ["enable-automation"])
options.add_experimental_option('useAutomationExtension', False)

service = Service(executable_path="chromedriver.exe")  # 경로 확인

driver = webdriver.Chrome(service=service, options=options)

# 자동화 감지 우회용 자바스크립트
driver.execute_cdp_cmd("Page.addScriptToEvaluateOnNewDocument", {
    "source": """
        Object.defineProperty(navigator, 'webdriver', {
            get: () => undefined
        })
    """
})

# 1. 네이버 검색
url = f"https://search.naver.com/search.naver?query={keyword}"
driver.get(url)

try:
    # 2. FAQ 더보기 버튼이 로딩될 때까지 대기
    faq_more = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((
            By.CSS_SELECTOR,
            "#main_pack > section.sc_new.sp_intent_faq > div > div > a"
        ))
    )

    # 3. 마우스 오버 후 클릭 (사람처럼)
    actions = ActionChains(driver)
    actions.move_to_element(faq_more).pause(1).click().perform()

    # 4. FAQ 페이지 로딩 대기
    WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CSS_SELECTOR, "strong"))
    )
    
    time.sleep(random.uniform(0.5, 1.5))  # 자연스러운 대기

    # 5. 질문 제목 출력 및 "?" 포함 항목만 따로 저장
    titles = driver.find_elements(By.CSS_SELECTOR, "strong")
    question_titles = []

    for title in titles:
        text = title.text
        # print(text)
        if "?" in text:
            question_titles.append(text)

    print("'?'가 포함된 질문 리스트:")
    for q in question_titles:
        lst.append(q)
        print("-", q)

except Exception as e:
    print("검색어에 등록된 FAQ가 없습니다.")

driver.quit()

df = pd.DataFrame({
    "질문": lst,
    "질문에 대한 대답": [""] * len(lst)  # 빈 문자열로 초기화
})

if lst == []:
    print("빈 리스트 이므로 csv파일로 저장하지 않았습니다.")
else:
    df.to_csv(f'{keyword} FAQ 질문.csv', index=False, encoding='utf-8-sig')
    print("FAQ 질문자료를 CSV 파일 저장 완료")
```

```python
# 질문: #container > div > div > div > div > section > div > ul > li:nth-child(i) > div > a

# 펼치기 아이콘: #container > div > div > div > div > section > div > ul > li:nth-child(i) > a > i.spib.ico_faq_fold

# 답변: #container > div > div > div > div > section > div > ul > li:nth-child(i) > div > div.answer_box

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import random
import pandas as pd

questions = []
answers = []

keyword = input("검색어를 입력해주세요: ")

# 크롬 옵션 설정
options = Options()
options.add_argument("--disable-gpu")
options.add_argument("start-maximized")
options.add_argument("disable-infobars")
options.add_argument("--disable-blink-features=AutomationControlled")
options.add_experimental_option("excludeSwitches", ["enable-automation"])
options.add_experimental_option('useAutomationExtension', False)

service = Service(executable_path="chromedriver.exe")
driver = webdriver.Chrome(service=service, options=options)

# 자동화 감지 우회
driver.execute_cdp_cmd("Page.addScriptToEvaluateOnNewDocument", {
    "source": """
        Object.defineProperty(navigator, 'webdriver', {
            get: () => undefined
        })
    """
})

# 1. 네이버 검색 페이지로 이동
url = f"https://search.naver.com/search.naver?query={keyword}"
driver.get(url)

try:
    # 2. FAQ 더보기 버튼 대기 후 클릭
    faq_more = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((
            By.CSS_SELECTOR,
            "#main_pack > section.sc_new.sp_intent_faq > div > div > a"
        ))
    )
    faq_more.click()

    # 3. FAQ 항목 1~20까지 반복 처리
    for i in range(1, 21):
        try:
            # 질문 요소 추출
            question_selector = f"strong"
            question_element = WebDriverWait(driver, 3).until(
                EC.presence_of_element_located((By.CSS_SELECTOR, question_selector))
            )
            question_text = question_element.text

            # 펼치기 버튼 클릭
            toggle_selector = f"#container > div > div > div > div > section > div > ul > li:nth-child({i}) > a > i.spib.ico_faq_fold"
            toggle_element = driver.find_element(By.CSS_SELECTOR, toggle_selector)
            driver.execute_script("arguments[0].click();", toggle_element)

            time.sleep(random.uniform(0.5, 1.2))  # 자연스러운 대기

            # 답변 추출
            answer_selector = f"#container > div > div > div > div > section > div > ul > li:nth-child({i}) > div > div"
            answer_element = WebDriverWait(driver, 3).until(
                EC.presence_of_element_located((By.CSS_SELECTOR, answer_selector))
            )
            answer_text = answer_element.text

            # 리스트에 추가
            questions.append(question_text)
            answers.append(answer_text)

            print(f"[{i}] Q: {question_text}")
            print(f"    A: {answer_text}\n")

        except Exception as inner_e:
            print(f"[{i}] FAQ 항목 로딩 실패 또는 비어있음")
            continue

except Exception as e:
    print("FAQ 더보기 버튼을 찾을 수 없거나 FAQ가 없습니다.")

driver.quit()

# # 데이터프레임 생성 및 저장
# df = pd.DataFrame({
#     "질문": questions,
#     "답변": answers
# })

# if df.empty:
#     print("FAQ 데이터가 없어 CSV로 저장하지 않습니다.")
# else:
#     df.to_csv(f'{keyword} FAQ_질문답변.csv', index=False, encoding='utf-8-sig')
#     print("CSV 파일로 저장 완료.")
```
